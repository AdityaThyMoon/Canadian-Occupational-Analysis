```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_excel("./Excel_Canada_Data.xlsx", skiprows=13)
df.rename(columns={"Occupation  ": "Occupation"}, inplace=True)

newfoundland_df = df[["Occupation", "Total", "Men", "Women"]].copy()

pei_df = df[["Occupation", "A1", "A2", "A3"]].copy()
pei_df.rename(columns={"A1": "Total", "A2": "Men", "A3": "Women"}, inplace=True)

nova_scotia_df = df[["Occupation", "B1", "B2", "B3"]].copy()
nova_scotia_df.rename(columns={"B1": "Total", "B2": "Men", "B3": "Women"}, inplace=True)

new_brunswick_df = df[["Occupation", "C1", "C2", "C3"]].copy()
new_brunswick_df.rename(columns={"C1": "Total", "C2": "Men", "C3": "Women"}, inplace=True)

quebec_df = df[["Occupation", "D1", "D2", "D3"]].copy()
quebec_df.rename(columns={"D1": "Total", "D2": "Men", "D3": "Women"}, inplace=True)

ontario_df = df[["Occupation", "E1", "E2", "E3"]].copy()
ontario_df.rename(columns={"E1": "Total", "E2": "Men", "E3": "Women"}, inplace=True)

manitoba_df = df[["Occupation", "F1", "F2", "F3"]].copy()
manitoba_df.rename(columns={"F1": "Total", "F2": "Men", "F3": "Women"}, inplace=True)

saskatchewan_df = df[["Occupation", "G1", "G2", "G3"]].copy()
saskatchewan_df.rename(columns={"G1": "Total", "G2": "Men", "G3": "Women"}, inplace=True)

alberta_df = df[["Occupation", "H1", "H2", "H3"]].copy()
alberta_df.rename(columns={"H1": "Total", "H2": "Men", "H3": "Women"}, inplace=True)

british_columbia_df = df[["Occupation", "I1", "I2", "I3"]].copy()
british_columbia_df.rename(columns={"I1": "Total", "I2": "Men", "I3": "Women"}, inplace=True)

yukon_df = df[["Occupation", "J1", "J2", "J3"]].copy()
yukon_df.rename(columns={"J1": "Total", "J2": "Men", "J3": "Women"}, inplace=True)

northwest_territories_df = df[["Occupation", "K1", "K2", "K3"]].copy()
northwest_territories_df.rename(columns={"K1": "Total", "K2": "Men", "K3": "Women"}, inplace=True)

nunavut_df = df[["Occupation", "L1", "L2", "L3"]].copy()
nunavut_df.rename(columns={"L1": "Total", "L2": "Men", "L3": "Women"}, inplace=True)
```

```python
def normalize_float_and_round(value, df):
  value = value / df[df["Occupation"] == "Total - Occupation - Unit group - National Occupational Classification (NOC) 2021 12"]["Total"].values[0]
  return round(float(value), 4)

def get_normalized_essential_workers(df):
  total_nurses = df[df["Occupation"] == "31301 Registered nurses and registered psychiatric nurses"]["Total"].values + df[df["Occupation"] == "31302 Nurse practitioners"]["Total"].values + df[df["Occupation"] == "32101 Licensed practical nurses"]["Total"].values
  total_police = df[df["Occupation"] == "42100 Police officers (except commissioned)"]["Total"].values + df[df["Occupation"] == "40040 Commissioned police officers and related occupations in public protection services"]["Total"].values
  total_firefighters = df[df["Occupation"] == "42101 Firefighters"]["Total"].values + df[df["Occupation"] == "40041 Fire chiefs and senior firefighting officers"]["Total"].values
  total_essential_workers_ontario = total_nurses + total_police + total_firefighters
  return normalize_float_and_round(total_essential_workers_ontario,df), normalize_float_and_round(total_nurses, df), normalize_float_and_round(total_police, df), normalize_float_and_round(total_firefighters, df)

alberta_total_count, alberta_nurses_count, alberta_police_count, alberta_firefighters_count = get_normalized_essential_workers(alberta_df)
british_columbia_total_count, british_columbia_nurses_count, british_columbia_police_count, british_columbia_firefighters_count = get_normalized_essential_workers(british_columbia_df)
manitoba_total_count, manitoba_nurses_count, manitoba_police_count, manitoba_firefighters_count = get_normalized_essential_workers(manitoba_df)
new_brunswick_total_count, new_brunswick_nurses_count, new_brunswick_police_count, new_brunswick_firefighters_count = get_normalized_essential_workers(new_brunswick_df)
newfoundland_total_count, newfoundland_nurses_count, newfoundland_police_count, newfoundland_firefighters_count = get_normalized_essential_workers(newfoundland_df)
northwest_territories_total_count, northwest_territories_nurses_count, northwest_territories_police_count, northwest_territories_firefighters_count = get_normalized_essential_workers(northwest_territories_df)
nova_scotia_total_count, nova_scotia_nurses_count, nova_scotia_police_count, nova_scotia_firefighters_count = get_normalized_essential_workers(nova_scotia_df)
nunavut_total_count, nunavut_nurses_count, nunavut_police_count, nunavut_firefighters_count = get_normalized_essential_workers(nunavut_df)
ontario_total_count, ontario_nurses_count, ontario_police_count, ontario_firefighters_count = get_normalized_essential_workers(ontario_df)
pei_total_count, pei_nurses_count, pei_police_count, pei_firefighters_count = get_normalized_essential_workers(pei_df)
quebec_total_count, quebec_nurses_count, quebec_police_count, quebec_firefighters_count = get_normalized_essential_workers(quebec_df)
saskatchewan_total_count, saskatchewan_nurses_count, saskatchewan_police_count, saskatchewan_firefighters_count = get_normalized_essential_workers(saskatchewan_df)
yukon_total_count, yukon_nurses_count, yukon_police_count, yukon_firefighters_count = get_normalized_essential_workers(yukon_df)

normalized_essential_workers_df = pd.DataFrame({
    "Province": ["Alberta", "British Columbia", "Manitoba", "New Brunswick", "Newfoundland and Labrador", "Northwest Territories", "Nova Scotia", "Nunavut", "Ontario", "Prince Edward Island", "Quebec", "Saskatchewan", "Yukon"],
    "Normalized_total_essential_workers_count": [alberta_total_count, british_columbia_total_count, manitoba_total_count, new_brunswick_total_count, newfoundland_total_count, northwest_territories_total_count, nova_scotia_total_count, nunavut_total_count, ontario_total_count, pei_total_count, quebec_total_count, saskatchewan_total_count, yukon_total_count],
    "Normalized_nurses_count": [
        alberta_nurses_count, british_columbia_nurses_count, manitoba_nurses_count, new_brunswick_nurses_count, newfoundland_nurses_count, northwest_territories_nurses_count,nova_scotia_nurses_count, nunavut_nurses_count, ontario_nurses_count,pei_nurses_count, quebec_nurses_count, saskatchewan_nurses_count, yukon_nurses_count],
    "Normalized_police_count": [alberta_police_count, british_columbia_police_count, manitoba_police_count, new_brunswick_police_count, newfoundland_police_count, northwest_territories_police_count,nova_scotia_police_count, nunavut_police_count, ontario_police_count,pei_police_count, quebec_police_count, saskatchewan_police_count, yukon_police_count],
    "Normalized_firefighters_count": [alberta_firefighters_count, british_columbia_firefighters_count, manitoba_firefighters_count,new_brunswick_firefighters_count, newfoundland_firefighters_count, northwest_territories_firefighters_count,nova_scotia_firefighters_count, nunavut_firefighters_count, ontario_firefighters_count,pei_firefighters_count, quebec_firefighters_count, saskatchewan_firefighters_count, yukon_firefighters_count]
})

normalized_essential_workers_df
```

```python
alberta_df["Province"] = "Alberta"
british_columbia_df["Province"] = "British Columbia"
manitoba_df["Province"] = "Manitoba"
new_brunswick_df["Province"] = "New Brunswick"
newfoundland_df["Province"] = "Newfoundland and Labrador"
northwest_territories_df["Province"] = "Northwest Territories"
nova_scotia_df["Province"] = "Nova Scotia"
nunavut_df["Province"] = "Nunavut"
ontario_df["Province"] = "Ontario"
pei_df["Province"] = "Prince Edward Island"
quebec_df["Province"] = "Quebec"
saskatchewan_df["Province"] = "Saskatchewan"
yukon_df["Province"] = "Yukon"

combined_df = pd.concat([alberta_df, british_columbia_df, manitoba_df, new_brunswick_df, newfoundland_df, northwest_territories_df, nova_scotia_df, nunavut_df, ontario_df, pei_df, quebec_df, saskatchewan_df, yukon_df], ignore_index=True)

combined_df_q2 = combined_df[combined_df["Occupation"].str.match(r"^\d{1}\s")]
combined_df_q2["Male Percentage"] = round((combined_df_q2["Men"] / combined_df_q2["Total"]),4)
combined_df_q2["Female Percentage"] = round((combined_df_q2["Women"] / combined_df_q2["Total"]),4)
combined_df_q2.shape
```

```python
import dash
from dash import html, dcc

app = dash.Dash("cp321")

app.layout = html.Div([
    html.H1("Final Project Dashboard", style={"color": "Black"}),
    html.Div([
    dcc.Dropdown(
        options=[
            {"label": "Police", "value": "Normalized_police_count"},
            {"label": "Nurses", "value": "Normalized_nurses_count"},
            {"label": "Firefighters", "value": "Normalized_firefighters_count"},
            {"label": "Total workers", "value": "Normalized_total_essential_workers_count"}
        ],
        id = "worker-dropdown",
        value="Normalized_total_essential_workers_count",
        clearable=False
    ),
    dcc.Graph(id="graph_1")
    ]),

    html.Div([

        dcc.RadioItems(
            id = "gender-radio",
            options = [
                {"label": "Male", "value": "male"},
                {"label": "Female", "value": "female"},
            ],
            value="male"
        ),
        dcc.Graph(id="graph_2")
    ]),

    html.Div([
        dcc.Checklist(
            id = "engineers-checklist",
            value = ["21311 Computer engineers (except software engineers and designers)", "21301 Mechanical engineers", "21310 Electrical and electronics engineers"],
            options = [
                {"label": "Mechanical Engineers", "value": "21301 Mechanical engineers"},
                {"label": "Computer Engineers","value": "21311 Computer engineers (except software engineers and designers)"},
                {"label": "Electrical and Electronics Engineers", "value": "21310 Electrical and electronics engineers"}
            ]
        ),
        dcc.Graph(id="graph_3")
    ]),

  html.Div([
    dcc.Slider(
        id = "number_slider",
        step=1,
        min=1,
        value=5,
        marks = {
            m : str(m)
            for m in range(1, 11)
        },
        max=10
    ),
    dcc.Dropdown(
        id="province-dropdown",
        clearable=False,
        value="Alberta",
        options=[{"label": province, "value": province} for province in combined_df["Province"].unique()]
    ),
    dcc.Graph(id="graph_4")
  ])
])
```

```python
# #geojson source: https://github.com/codeforgermany/click_that_hood/blob/main/public/data/canada.geojson

from dash.dependencies import Input, Output

@app.callback(
    Output("graph_1", "figure"),
    Input("worker-dropdown", "value")
)

def update_graph(job):
  updated_graph = px.choropleth(
    normalized_essential_workers_df,
    geojson="https://raw.githubusercontent.com/codeforgermany/click_that_hood/main/public/data/canada.geojson",
    locations="Province",
    locationmode="geojson-id",
    featureidkey="properties.name",
    color=job,
    color_continuous_scale="Reds_r",
    title=f"Normalized number of essential workers per province in Canada (2021 consensus)"
  )
  updated_graph.update_geos(fitbounds="locations", visible=False)
  return updated_graph

@app.callback(
    Output("graph_2", "figure"),
    Input("gender-radio", "value")
)

def update_graph_2(gender_radio):
  if(gender_radio == "male"):
    column = "Male Percentage"
  elif(gender_radio == "female"):
    column = "Female Percentage"

  updated_graph_2 = px.density_heatmap(
      combined_df_q2,
      x="Occupation",
      y = "Province",
      z=column,
      color_continuous_scale="Blues",
      text_auto=".1%",
      title = f"Percentage of workers who are {column.split()[0]} per industry, based on administrative units in Canada"
  )

  updated_graph_2.update_layout(
    xaxis_tickangle=-45,
    margin=dict(t=60, b=150),
    xaxis_title="Occupation",
    yaxis_title="Province",
    height=1000,
  )

  return updated_graph_2

@app.callback(
    Output("graph_3", "figure"),
    Input("engineers-checklist", "value")
)

def update_graph_3(engineer):
  engineer_df = combined_df[combined_df["Occupation"].isin(engineer)]
  updated_graph_3 = px.bar(
      engineer_df,
      color="Occupation",
      barmode="stack",
      title = "Number of engineeers by province in Canada",
      x = "Province",
      y = "Total",
  )

  updated_graph_3.update_layout(
    xaxis_tickangle=-45,
    xaxis_title="Occupation",
    yaxis_title="Total engineers",
    height=600,
    margin=dict(t=60, b=150),
  )

  return updated_graph_3

@app.callback(
    Output("graph_4", "figure"),
    Input("number_slider", "value"),
    Input("province-dropdown", "value")
)

def update_graph_4(number_slider, dropdown):
  temp_df = combined_df.copy()
  temp_df = temp_df[~temp_df["Occupation"].isin(["Occupation - not applicable 13", "Total - Occupation - Unit group - National Occupational Classification (NOC) 2021 12"])]
  temp_df = temp_df[temp_df["Province"] == dropdown]
  temp_df["Total_normalized"] = temp_df["Total"].apply(lambda x: normalize_float_and_round(x, combined_df))
  top_occupations_df = temp_df.sort_values(by="Total_normalized", ascending=False).groupby("Province").head(number_slider)

  graph_4 = px.sunburst(
      top_occupations_df,
      path=["Province", "Occupation"],
      values="Total_normalized",
      title=f"Top {number_slider} occupations per province in Canada (normalized)",
      color="Total_normalized",
      color_continuous_scale="Greens"
  )

  graph_4.update_layout(
      width=1300,
      height=1300,
      uniformtext=dict(minsize=5, mode="hide"),
  )

  return graph_4

app.run(debug=True)
```
